```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x_pred):
    return x_pred * (1 - x_pred)

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [0]])

np.random.seed(0)

weights = np.random.rand(2, 1) # Adjusted for input shape
bias = np.random.rand(1, 1)
learning_rate = 0.1
epochs = 1000
batch_size = 2

for epoch in range(epochs):
    indices = np.random.permutation(len(x))
    X_shuffled = x[indices]
    y_shuffled = y[indices]

    for i in range(0, len(x), batch_size):
        X_batch = X_shuffled[i : i + batch_size]
        y_batch = y_shuffled[i : i + batch_size]

        z = np.dot(X_batch, weights) + bias
        y_pred = sigmoid(z)

        error = y_batch - y_pred

        d_pred = error * sigmoid_derivative(y_pred)
        d_weights = np.dot(X_batch.T, d_pred)
        d_bias = np.sum(d_pred, axis=0, keepdims=True)

        weights += learning_rate * d_weights
        bias += learning_rate * d_bias

print("Predicted output")
print(sigmoid(np.dot(x, weights) + bias))
```
