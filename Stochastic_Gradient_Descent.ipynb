```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_grad(s):
    return s * (1 - s)

x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0], [0], [0],[1]])

np.random.seed(0)

weights = np.random.rand(2,1)
bias = np.random.rand(1,1)

learning_rate = 0.1
epochs = 1000

for _ in range(epochs):
    for i in range(len(x)):
        x_i = x[i].reshape(1,-1)
        y_i = y[i].reshape(1,-1)

        z = np.dot(x_i, weights) + bias
        y_pred = sigmoid(z)

        error = y_i - y_pred

        d_pred = error * sigmoid_grad(y_pred)

        d_weights = np.dot(x_i.T, d_pred)
        d_bias = d_pred

        weights += learning_rate * d_weights
        bias += learning_rate * d_bias

print("Predicted output:")
print(sigmoid(np.dot(x, weights) + bias))
```
