```python
import numpy as np

x1 = np.array([0, 0, 1, 1])
x2 = np.array([0, 1, 0, 1]) # Corrected typo from [0.1.0.1]
t = np.array([0, 1, 1, 0])  # Corrected typo from [0,1,1]: and added a fourth element for consistency with x1, x2

m = len(t)

max_iter = 4000  # Corrected variable name from max-iter
learn_rate = 0.1 # Corrected variable name from learn rate

w1 = np.random.uniform(-1, 1) * 0.2
w2 = np.random.uniform(-1, 1) * 0.2
b = 0

def sigmoid(z):  # Corrected parameter name from 2 to z
    return 1 / (1 + np.exp(-z)) # Corrected sigmoid function syntax

def sigmoid_derivative(s): # Corrected function name from sigmoid-derivative and removed unmatched quote
    return s * (1 - s) # Corrected syntax from sx(1-5)

for _ in range(max_iter): # Corrected variable name from max-iter and added loop variable _
    # Initialize gradients for batch update
    dw1 = 0
    dw2 = 0 # Corrected variable name from d002:0
    db = 0

    for i in range(m):
        z = x1[i] * w1 + x2[i] * w2 + b # Corrected variable names, operators, and index for x2
        y = sigmoid(z) # Corrected variable name from 4 to y and parameter from 2 to z
        error = t[i] - y # Corrected variable name from error: to error = and Cyrillic У to y
        d = error * sigmoid_derivative(y) # Corrected missing operator and function name

        dw1 += d * x1[i] # Accumulate gradients
        dw2 += d * x2[i] # Accumulate gradients
        db += d # Accumulate gradients

    # Update weights after iterating through all samples (batch update)
    w1 -= learn_rate * (dw1 / m) # Corrected variable name, operator, and assignment
    w2 -= learn_rate * (dw2 / m) # Corrected variable name, operator, and assignment
    b -= learn_rate * (db / m) # Corrected variable name, operator, and assignment

print("Final weights and bias:")
print("w1 = ", w1) # Corrected variable name from nol to w1 and removed trailing syntax
print("w2 = ", w2) # Corrected variable name from W2 to w2
print("b = ", b) # Corrected trailing syntax

print("\nPredictions:") # Corrected typo In Fredictions to Predictions
for i in range(m):
    z = x1[i] * w1 + x2[i] * w2 + b # Corrected variable names and operators
    y = sigmoid(z) # Corrected variable name from 4 to y and parameter from 2 to z
    print(x1[i], x2[i], "→", round(y))
```
