```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    # Assuming x is the output of the sigmoid function
    return x * (1 - x)

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [1]])

np.random.seed(0)

weights = np.random.rand(2, 1) # Corrected from 2.1 to (2, 1)
bias = np.random.rand(1)

v_w = np.zeros_like(weights) # Corrected u_w to v_w and zeros-like to zeros_like
v_b = np.zeros_like(bias)   # Corrected v-b and zeros-like to zeros_like

learning_rate = 0.1 # Corrected learning-rate to learning_rate
beta = 0.9

epochs = 1000 # Renamed 'epoch' to 'epochs' for clarity with loop
batch_size = 2 # Corrected batch-size to batch_size

for _ in range(epochs): # Corrected for-in and used epochs
    indices = np.random.permutation(len(x)) # Corrected Clen
    X_shuffled = x[indices] # Corrected X-Shuffled and Cindices
    y_shuffled = y[indices] # Corrected 4-shuffled and Cindices

    for i in range(0, len(x), batch_size): # Corrected o and !
        X_batch = X_shuffled[i:i + batch_size] # Corrected X-batch and Cisit batch-sizej
        y_batch = y_shuffled[i:i + batch_size] # Corrected 4-batch and it

        z = np.dot(X_batch, weights) + bias # Corrected 2, &, and x-batch
        y_pred = sigmoid(z) # Corrected 4-pred and variable name

        error = y_batch - y_pred # Corrected missing operator
        d_pred = error * sigmoid_derivative(y_pred) # Corrected missing operator and function name

        # Correct calculation of gradients assuming backpropagation
        grad_w = np.dot(X_batch.T, d_pred) # Corrected grad-w, T. and x-batch
        grad_b = np.sum(d_pred, axis=0) # Corrected grad-b and np-sum

        # Update momentum variables
        v_w = beta * v_w + learning_rate * grad_w # Corrected v-w, betaxv-w, learning-rate, grad-w
        v_b = beta * v_b + learning_rate * grad_b # Corrected v-b, learning-rate, grad-b

        # Apply updates
        weights += v_w # Corrected v_W to v_w
        bias += v_b # Corrected v-b to v_b

print("Predicted output:")
print(sigmoid(np.dot(x, weights) + bias)) # Corrected tbias to + bias
```
