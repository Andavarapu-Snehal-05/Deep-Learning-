```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

X = np.linspace(0, 10, 50).reshape(-1, 1)
Y = 4 * X + 2 + np.random.randn(50, 1)

w = np.random.randn(1, 1) # Initialize weight
b = np.random.randn(1, 1) # Initialize bias

learning_rate = 0.01
epochs = 1000

lambda_reg = 0.1 # L1 regularization parameter

n = len(X)

losses = []

for epoch in range(epochs):
    y_pred = np.dot(X, w) + b

    mse = np.mean((Y - y_pred) ** 2)
    l1_term = lambda_reg * np.sum(np.abs(w))
    loss = mse + l1_term
    losses.append(loss)

    # Gradients
    dw = (-2/n) * np.dot(X.T, (Y - y_pred)) + lambda_reg * np.sign(w)
    db = (-2/n) * np.sum(Y - y_pred)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

print("Final weight:", w)
print("Final Bias:", b)

plt.scatter(X, Y, label="Actual data")
plt.plot(X, np.dot(X, w) + b, color='red',
         label="LASSO REGRESSION FIT")
plt.legend()
plt.title("LASSO Regression using Neural Network")
plt.show()

plt.plot(losses)
plt.ylabel("Training Loss")
plt.xlabel("Epochs")
plt.show()
```
