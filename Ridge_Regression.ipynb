```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0.10, 50, 50).reshape(-1, 1)
y = 3 * x + 5 + np.random.randn(50, 1)

np.random.seed(0)
w = np.random.randn(1, 1) # Changed to (1, 1) to ensure consistent shape
learning_rate = 0.00001 # Reduced learning rate to prevent overflow
epochs = 1000

b = 0.0

lambda_reg = 0.1

n = len(x)

losses = []

for epoch in range(epochs):

    y_pred = np.dot(x, w) + b

    mse = np.mean((y - y_pred)**2)

    l2_loss = lambda_reg * np.sum(w**2)

    loss = mse + l2_loss

    losses.append(loss)

    dw = (-2/n) * np.dot(x.T, (y - y_pred)) + 2 * lambda_reg * w

    db = (-2/n) * np.sum(y - y_pred)

    w -= learning_rate * dw

    b -= learning_rate * db

print("Final weight:", w)
print("Final bias:", b)

plt.scatter(x, y, label="Actual data")

plt.plot(x, np.dot(x, w) + b, color='red', label="Ridge Regression fit")

plt.legend()
plt.title("Ridge Regression using Neural Network")
plt.show()

plt.plot(losses)

plt.title("Training Loss")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
