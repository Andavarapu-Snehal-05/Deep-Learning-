```python
import numpy as np

x1 = np.array ([0,0,1,1])
x2 = np.array ([0,1,0,1])
t= np.array ([0,1,1,0]) # Corrected array definition
m= len(t)

max_iter = 90000 # Corrected variable name

learn_rate = 0.3 # Corrected variable name

w1 = np.random.uniform (-1,1) *0.2
w2 = np.random.uniform (-1,1) *0.2 # Corrected typo
w3 = np.random.uniform (-1,1)*0.2
w4 = np.random.uniform (-1,1)*0.2 # Corrected variable name
w5 = np.random.uniform (-1,1)*0.2 # Corrected variable name
w6 = np.random.uniform (-1,1)*0.2 # Corrected variable name

b1 = 0.0 # Corrected variable name and assignment
b2 = 0.0 # Corrected variable name
b3 = 0.0 # Corrected variable name

def sigmoid (z):
    return 1/(1+ np.exp(-z)) # Corrected function name and syntax

def sigmoid_derivative (s):
    return s* (1-s) # Corrected function name and variable

pre_mse = float('inf') # Corrected variable name


for _ in range (max_iter): # Corrected loop variable and variable name

    # Initializing derivatives for this iteration
    dw1 = dw2 = dw3 = dw4 = dw5 = dw6 = 0.0
    db1 = db2 = db3 = 0.0

    mse = 0.0

    for i in range (m):

        # Forward pass
        zh1 = x1[i] * w1 + x2[i] * w2 + b1 # Corrected variable names
        zh2 = x1[i] * w3 + x2[i] * w4 + b2 # Corrected variable names

        h1 = sigmoid (zh1) # Corrected variable name
        h2 = sigmoid (zh2)

        z0 = h1 * w5 + h2 * w6 + b3 # Corrected variable names
        y = sigmoid (z0)

        error = y - t[i] # Corrected variable assignment

        mse += (t[i] - y) ** 2 # Corrected calculation

        # Backward pass
        d_out = error * sigmoid_derivative(y) # Corrected variable name

        dw5 += d_out * h1
        dw6 += d_out * h2
        db3 += d_out

        dh1 = d_out * w5 * sigmoid_derivative(h1) # Corrected variable name
        dh2 = d_out * w6 * sigmoid_derivative(h2) # Corrected variable name

        dw1 += dh1 * x1[i]
        dw2 += dh1 * x2[i]
        dw3 += dh2 * x1[i]
        dw4 += dh2 * x2[i]

        db1 += dh1 * 1.0
        db2 += dh2 * 1.0

    # Update weights and biases
    w1 -= learn_rate * (dw1 / m) # Corrected variable names and syntax
    w2 -= learn_rate * (dw2 / m) # Corrected variable names
    w3 -= learn_rate * (dw3 / m) # Corrected variable names
    w4 -= learn_rate * (dw4 / m) # Corrected variable names
    w5 -= learn_rate * (dw5 / m) # Corrected variable names
    w6 -= learn_rate * (dw6 / m) # Corrected variable names

    b1 -= learn_rate * (db1 / m) # Corrected variable names
    b2 -= learn_rate * (db2 / m) # Corrected variable names
    b3 -= learn_rate * (db3 / m) # Corrected variable names

    mse /= m



    if (abs (mse - pre_mse) < 0.0001):
        break

    pre_mse = mse # Corrected variable name

print ("Final weights:")
print (w1, w2, w3, w4, w5, w6) # Corrected variable names
print ("Biases:", b1, b2, b3) # Corrected variable names

print("In Predictions:") # Corrected string

for j in range (m):

    h1 = sigmoid (x1[j] * w1 + x2[j] * w2 + b1) # Corrected variable names and syntax
    h2 = sigmoid (x1[j] * w3 + x2[j] * w4 + b2) # Corrected variable names and syntax

    y = sigmoid (h1 * w5 + h2 * w6 + b3) # Corrected variable names

    print (x1[j], x2[j], "->", y) # Corrected loop variable and output
```
