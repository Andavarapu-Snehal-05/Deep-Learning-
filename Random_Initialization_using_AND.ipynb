```python
import numpy as np

x1 = np.array([0,0,1,13])

x2= np.array ([0,1,0,1])

t= np.array ([0,0,0,1])

m = len(t)

max_iter=4000

learn_rate = 0.1

W1 = np.random.uniform(-1, 1) + 0.2

W2= np.random.uniform (-1,1)*0.2

b=0 # Assuming this was intended to initialize bias 'b'

def sigmoid(a):
    return 1 / (1 + np.exp(-a))

def sigmoid_derivative(s):
    return s * (1 - s)

for _ in range (max_iter): # Corrected loop syntax and variable name
    # Removed '100' and 'rarSPEC=' as they were syntax errors or incomplete statements

    dw1=0
    dw2=0 # Corrected typo from duo2
    db=0

    for i in range (m):
        # Removed 'OF'
        z = x1[i] * W1 + x2[i] * W2 + b # Corrected variable names and added multiplication
        y = sigmoid(z) # Corrected function call
        error = t[i] - y # Corrected variable name
        d = error * sigmoid_derivative(y) # Added multiplication and corrected function call
        dw1 += d * x1[i] # Corrected variable name, multiplication, and accumulation
        dw2 += d * x2[i] # Corrected variable name, multiplication, and accumulation
        db += d # Corrected accumulation

    W1 += learn_rate * (dw1 / m) # Corrected variable names and division
    W2 += learn_rate * (dw2 / m) # Corrected variable names and division
    b += learn_rate * (db / m) # Corrected variable names and division

print("Final weights and bias:") # Corrected print statement
print("W1 = ", W1) # Corrected variable name
print("W2 = ", W2) # Corrected variable name
# Removed '00'
print("b=", b)

print("In Prediction:") # Corrected print statement

for i in range(m):
    z = x1[i] * W1 + x2[i] * W2 + b # Corrected variable names and added multiplication
    y = sigmoid(z) # Corrected function call and case sensitivity
    print(x1[i], x2[i],"→", round (y))
```
